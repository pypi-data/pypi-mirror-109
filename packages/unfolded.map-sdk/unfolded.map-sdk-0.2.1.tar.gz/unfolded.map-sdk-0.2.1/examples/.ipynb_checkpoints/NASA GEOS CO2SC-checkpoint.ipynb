```python
import pandas as pd
import xarray as xr
import h3
dates = ['20210331_0130','20210331_0430','20210331_0730','20210331_1030','20210331_1330','20210331_1630','20210331_1930']
path= '/Users/ilya/Data/NASA/'
h3_resolution=4
```

```python
df1 = xr.open_dataset(path+'GEOS.fp.asm.tavg3_2d_chm_Nx.20210331_0130.V01.nc4').to_dataframe()
```

```python
df1
```

```python
def aggregate_h3(df):
    gb1h3 = df[['CO2SC']].groupby([lambda r: h3.geo_to_h3(r[0],r[1],h3_resolution),'time'])
    df1h3 = gb1h3.agg({'CO2SC':['median']})
    df1h3.columns = df1h3.columns.droplevel(1)
    return df1h3
```

```python
aggregate_h3(df1).to_csv(path+'df1h3.csv',index_label=['h3','time'])
```

```python
df = pd.concat(
    aggregate_h3(xr.open_dataset(path+'GEOS.fp.asm.tavg3_2d_chm_Nx.'+d+'.V01.nc4')
    .to_dataframe()[["CO2SC"]]) for d in dates
)
df.to_csv(path+'GEOS.20210331.h3-'+str(h3_resolution)+'.csv',index_label=['h3','time'])
```
